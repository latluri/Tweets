```python
import pandas as pd
#since_dt='2018-09-04'
tweets_collected=pd.read_csv('tweets_ford_%s.csv'%output_name,skip_blank_lines=True)

#tweets_collected=pd.read_csv('tweets_ford_%s.csv'%since_dt,skip_blank_lines=True,skiprows=2)
#tweets_collected=tweets_collected.dropna(subset=['tweet.user.name'])
#tweets_collected=tweets_collected.reset_index()
```

```python
import re
import emoji

train=tweets_collected[['tweet.user.name','tweet.text.string']]
train["tweet.text.string"]=train["tweet.text.string"].apply(lambda x: emoji.demojize(str(x)))
```

```python

train["tweet_text"]=train['tweet.text.string'].str.replace('https?://[A-Za-z0-9./]+','').str.replace('@Ford','Q1&6s5%&8bhgs40qo5ophf').str.replace('@[A-Za-z0-9_]+','').str.replace('Q1&6s5%&8bhgs40qo5ophf','@Ford')
train["tweet_text"]=train["tweet_text"].str.replace('.','').str.replace('!','').str.replace("n't",' not')
```

```python
train["tweet_text"]=train["tweet_text"].str.lower()
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
stop.remove('and')
stop.remove('for')
stop.remove('not')
```

```python
train['tweet_text']=train['tweet_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
car=list(pd.read_csv("car_jargon.txt",index_col=False).Entity.values)
#d=pd.concat([car, car], ignore_index=True,axis=1)
#d[0].map(str) + " " + d[1]
train["Car_related"]=train.tweet_text.apply(lambda x: ' '.join([word for word in x.split() if word in (car)]))
```

```python
ford_models=list(pd.read_csv("ford_models.txt",index_col=False).Entity.values)
train["Ford_Model"]=train.tweet_text.apply(lambda x: ' '.join([word for word in x.split() if word in (ford_models)]))
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
pos=[]
neg=[]
neu=[]
comp=[]

for sentence in train['tweet.text.string']:
     print    
     ss = sid.polarity_scores(sentence)
     pos.append(ss['pos'])
     neg.append(ss['neg'])
     neu.append(ss['neu'])
     comp.append(ss['compound'])
```

```python
train["pos"]=pos
train["neg"]=neg
train["neu"]=neu
train["comp"]=comp
```

```python
final=pd.merge(tweets_collected,train.loc[:,['tweet_text', 'Car_related','Ford_Model', 'pos', 'neg', 'neu', 'comp']], left_index=True, right_index=True)
```

```python
final
```

```python
final.loc[final['tweet.user.followers_count'] >= 10000, 'Influencer'] = 1
final.loc[final['tweet.user.followers_count'] < 10000, 'Influencer'] = 0
final.loc[final['tweet.retweet_count'] < 100, 'trending_v1'] = 0
final.loc[final['tweet.retweet_count'] >= 100, 'trending_v1'] = 1
```

```python
final.to_csv('ford_final_%s.csv'%output_name, sep=',', encoding='utf-8',index=False)
#final.to_csv('ford_final_2019-03-14.csv', sep=',', encoding='utf-8',index=False)
```
